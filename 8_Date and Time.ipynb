# Calendar Module

The calendar module allows you to output calendars and provides additional useful functions for them

#### class calendar.TextCalendar([firstweekday])

This class can be used to generate plain text calendars.

```python
>>> import calendar
>>> 
>>> print calendar.TextCalendar(firstweekday=6).formatyear(2015)
```


To learn more about different calendar functions, click [here](https://docs.python.org/2/library/calendar.html#calendar.setfirstweekday).

You are given a date. Your task is to find what the day is on that date.
- Input format : MM DD YYYY

````python
import calendar

mm, dd, yyyy = map(int, input().split())

print(calendar.day_name[calendar.weekday(yyyy, mm, dd)].upper())
````

# Time Delta

You are given 2 timestamps in the format given below:

Day dd Mon yyyy hh:mm:ss +xxxx

Here +xxxx represents the time zone. Your task is to print the absolute difference (in seconds) between them.

- Input Format : The first line contains $T$, the number of testcases. Each testcase contains $2$ lines, representing time $t_1$ and time $t_2$.

cf : https://docs.python.org/2/library/datetime.html

````python
#!/bin/python3

import sys
from datetime import datetime as dt

def time_delta(t1, t2):
    # Complete this function
    return int(abs(t1 - t2).total_seconds())

if __name__ == "__main__":
    t = int(input().strip())
    
    fmt = '%a %d %b %Y %H:%M:%S %z'
    
    for a0 in range(t):
        t1 = dt.strptime(input(), fmt)
        t2 = dt.strptime(input(), fmt)
        delta = time_delta(t1, t2)
        print(delta)
````
